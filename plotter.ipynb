```python
import sys, subprocess
from matplotlib import pyplot as plt

sizes=[]
accuracy=[]
x_values=list(range(1,101))
for i in x_values:
    result=subprocess.check_output(['python3', 'compression.py', str(i)]).decode('utf-8').split()
    sizes.append(int(result[0]))
    accuracy.append(float(result[1]))

```

```python
figure, ax=plt.subplots()
ax.plot(x_values, accuracy)
ax2=ax.twinx()
ax2.plot(x_values, sizes)
plt.show()
```
